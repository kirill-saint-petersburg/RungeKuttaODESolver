```python
import numpy as np

Data = np.loadtxt('result19_05_2020_23_58_47.out')
```

```python
phi = [np.arctan2(item[3], item[2]) for item in Data]

hist, bins = np.histogram(phi, bins=128)

width = 0.8 * (bins[1] - bins[0])

center = (bins[:-1] + bins[1:]) * 0.5
```

```python
import matplotlib.pyplot as plt

#plt.bar(center, hist, align='center', width=width)

plt.plot(center / np.pi, [(hist[i] * (np.sin(center[i] * 0.5) ** 2)) for i in range(center.size)])
```
